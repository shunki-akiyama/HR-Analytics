## Feature Description
Employee Name:	           Employee’s full name<br>
EmpID:	                   Employee ID is unique to each employee<br>
MarriedID:	               Is the person married (1 or 0 for yes or no)<br>
MaritalStatusID:           Marital status code that matches the text field MaritalDesc<br>
EmpStatusID:               Employment status code that matches text field EmploymentStatus<br>
DeptID:	                   Department ID code that matches the department the employee works in<br>
PerfScoreID:               Performance Score code that matches the employee’s most recent performance score<br>
FromDiversityJobFairID:    Was the employee sourced from the Diversity job fair? 1 or 0 for yes or no<br>
Salary:	                   The person’s anualu pay. <br>
Termd:	                   Has this employee been terminated - 1 or 0<br>
PositionID:	               An integer indicating the person’s position<br>
Position:	               The text name/title of the position the person has<br>
State:	                   The state that the person lives in<br>
Zip:	                   The zip code for the employee<br>
DOB:	                   Date of Birth for the employee<br>
Sex:	                   Sex - M or F<br>
MaritalDesc:	           The marital status of the person (divorced, single, widowed, separated, etc)<br>
CitizenDesc:	           Label for whether the person is a Citizen or Eligible NonCitizen<br>
HispanicLatino:	           Yes or No field for whether the employee is Hispanic/Latino<br>
RaceDesc:	               Description/text of the race the person identifies with<br>
DateofHire:	               Date the person was hired<br>
DateofTermination:	       Date the person was terminated, only populated if, in fact, Termd = 1<br>
TermReason:	               A text reason / description for why the person was terminated<br>
EmploymentStatus:          A description/category of the person’s employment status. <br>

Department:	               Name of the department that the person works in<br>
ManagerName:	           The name of the person’s immediate manager<br>
ManagerID:	               A unique identifier for each manager.<br>

EmpSatisfaction:           A basic satisfaction score between 1 and 5, as reported on a recent employee satisfaction survey<br>
SpecialProjectsCount:	   The number of special projects that the employee worked on during the last 6 months<br>
LastPerformanceReviewDate: The most recent date of the person’s last performance review.<br>

### Create a  variable
#employmentperiod

```python
from datetime import datetime
datadate = datetime.strptime('2019-02-28', '%Y-%m-%d')
```

```python
def employmentperiod(x):
    if x.Termd == 0:
        period = (datadate - x.DateofHire) / 365
    else:
        period = (x.DateofTermination - x.DateofHire) / 365
    return period
```

```python
df['EmpPeriod'] = df.apply(lambda x: employmentperiod(x), axis=1)
df.head()
```

```python
timedelta = datetime.strptime('2019-02-28', '%Y-%m-%d') - datetime.strptime('2019-02-27', '%Y-%m-%d')
df['EmpPeriod'] = df['EmpPeriod'] / timedelta
```

### Preprocess val&test data

```python
def preprocess_classif_data(X, y):
    
    #One-hot coding (Category)
    temporary_one_hot = pd.get_dummies(X[category_col])
    
    # We align the columns by adding any missing column to X.
    _, one_hot_values = X_one_hot.align(temporary_one_hot, 'left', axis=1)
    one_hot_values.fillna(0, inplace=True)
  
    
    #Z-score (Continuous value)
    X_zscored_continuous_variable = (X[continuous_col] - continuous_mean)/ continuous_std
    
    #Minimun (Ordinal value)
    preprocessed_ordinal_value = X[ordinal_col] - ordinal_min
    
    #Preprocessing date
    df_temp = pd.DataFrame()
    df_temp['month_DOB'] = X["DOB"].apply(lambda row: row.month)
    df_temp['year_DOB'] = X["DOB"].apply(lambda row: row.year)
    df_temp['day_DOB'] = X["DOB"].apply(lambda row: row.day)

    df_temp['month_DateofHire'] = X["DateofHire"].apply(lambda row: row.month)
    df_temp['year_DateofHire'] = X["DateofHire"].apply(lambda row: row.year)
    df_temp['day_DateofHire'] = X["DateofHire"].apply(lambda row: row.day)

    X_preprocess_date = pd.concat([df_temp["month_DOB"],df_temp["year_DOB"],df_temp["day_DOB"],df_temp["month_DateofHire"],
                               df_temp["year_DateofHire"], df_temp["day_DateofHire"]], axis=1)
    
    #Concate
    final_X = pd.concat([preprocessed_ordinal_value, X_preprocess_date,
                           X_zscored_continuous_variable, one_hot_values], axis = 1)
    final_y = y
    
    #Scaling (Log normalization)
    final_X["year_DOB"] = np.log1p(final_X["year_DOB"])
    final_X["year_DateofHire"] = np.log1p(final_X["year_DateofHire"])
        
    return final_X, final_y
```

```python
final_X_test, final_y_test = preprocess_classif_data(X_test, y_test)
final_X_val, final_y_val = preprocess_classif_data(X_val, y_val)
```

```python
# CV on number of layer 
def create_custom_model(input_dim, output_dim, nodes, n=1, name='model'):
    def create_model():
        model = Sequential(name=name)
        for i in range(n):
            model.add(Dense(nodes, input_dim=input_dim, activation='relu'))
            model.add(Dropout(0.2))
            
        model.add(Dense(output_dim, activation='sigmoid'))
        model.compile(loss='binary_crossentropy', 
                      optimizer=Adam(0.01), 
                      metrics=['accuracy',tf.keras.metrics.AUC()])
        #print(model.summary())
        return model
    return create_model

np.random.seed(52)
models = [create_custom_model(n_features, n_classes, 16, i, f'model_{i}') for i in range(1,10)]

for create_model in models:
    print(create_model().summary())
```
